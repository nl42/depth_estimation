# Laws' Mask

[Based on](https://courses.cs.washington.edu/courses/cse455/09wi/Lects/lect12.pdf)

```python
test = lambda x, y=3 : x+y
test(1)
```

```python
def mask(image, first, second=None):
    filters = {
                level    : [1, 4, 6, 4, 2],
                edge     : [-1, -2, 0, 2, 1],
                spot     : [-1, 0, 2, 0, -1],
                ripple   : [1, -4, 6, -4, 1]
              } 
    
    if second is None:
        return cv2.filter2D(y, filters[first]*filters[first])
    else:
        return np.mean(cv2.filter2D(y, -1, filters[first]*filters[second]), 
                       cv2.filter2D(y, -1, filters[second]*filters[first]))

def Laws(ycc):
    y, cr, cb = cv2.split(ycc_0)
    level, edge, spot, ripple = 'level', 'edge', 'spot', 'r'
    
    
    masks = [
                mask(y, level, edge),
                mask(y, level, ripple),
                mask(y, edge, spot),
                mask(y, spot),
                mask(y, ripple),
                mask(y, level, spot),
                mask(y, edge),
                mask(y, edge, ripple),
                mask(y, spot, ripple),
                mask(cr, level, edge),
                mask(cb, level, edge)
             ]
    
    return masks
```
